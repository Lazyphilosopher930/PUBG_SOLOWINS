```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error
```

```python
player=pd.read_csv("PUBG_Player_Statistics.csv")
```

```python
feature=['solo_WinRatio','solo_KillDeathRatio','solo_TeamKills','solo_HeadshotKillRatio','solo_HeadshotKills','solo_Rating','solo_LongestKill','solo_DamagePg','solo_Boosts','solo_DamageDealt','solo_RoundsPlayed','solo_WinTop10Ratio','solo_Top10s','solo_Top10Ratio']
x=player[feature]
y=player.solo_Wins
print(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.1) 
```

```python
linreg = LinearRegression()
linreg.fit(x_train, y_train)
```

```python
linreg.score(x_test,y_test)
```
